# AMA Supercross - Seattle Lumen Field 2026

**Date Started**: 2026-01-10

## Intro

### Domain Problem

*Who will take Top 3 for the Supercross 450 class event in Seattle?*

The objective of this project is to predict the Top 3 podium finishers for the **450 Supercross class** at the Seattle round taking place at **Lumen Field** on February 14th. 

Supercross racing is highly stochastic due to factors such as rider crashes, start position variability, and rapidly changing track conditions, making exact predictions unreliable. As a result, this project approaches the problem mathematically by simulating race outcomes multiple times and evaluating the frequency with which riders finish on the podium.

### Modeling Approach

To generate realistic podium predictions, this analysis combines historical rider performance data with Seattle-specific track and environmental factors, then uses a *Monte Carlo* simulation framework to model race variability. Instead of predicting a single finishing order, the model simulates the race 500+ times, allowing uncertainty and race-day randomness to be explicitly incorporated.

Each simulation produces a full finishing order, from which podium placements are recorded. Final predictions are based on the empirical distribution of podium finishes across all simulations (most counted podiums per rider).

##### What is a *Monte Carlo* simulation?

- A Monte Carlo simulation is a computer-based technique that uses repeated random sampling to model the probability of different outcomes in complex situations with many uncertain variables, helping to understand risk and forecast potential results. 

- Named after the Monaco casino for its reliance on chance, it runs thousands of scenarios by plugging in random inputs from defined probability distributions to generate a range of possible results, offering insights beyond simple linear projections. 

#### Variables of Interest

**Base Performance:** Each rider's overall series performance from previous six seasons.
- This represents a rider’s long-term competitive strength and is estimated using their historical results across the previous six Supercross seasons.
- This includes average finishing position, points earned per race, and overall consistency.
- This variable provides a stable baseline for rider ability and prevents the model from overreacting to short-term fluctuations.

**Recent Form:** Each rider's overall series performance from the 2025 season.
- This will capture short-term form and momentum by focusing on results from the 2025 Supercross season.
- Riders may experience changes in speed due to bike setup improvements, team switches, fitness changes, or confidence gained from recent results.
- This variable allows the model to adjust rider expectations dynamically as the season progresses.

**Seattle Track Performance:** Each rider's performance on Seattle tracks in previous six seasons.
- Some riders historically perform better or worse at specific venues due to soil composition, stadium layout, or rhythm section design.
- This measures how each rider has performed in prior Seattle Supercross events over the past six seasons, providing a venue-specific adjustment to baseline performance.

**DNF/Crash Rate:** Each rider's overall DNFs and major crashes from previous six seasons.
- This variable quantifies each rider’s historical frequency of DNFs and major crashes.
- Since a single crash can immediately remove a rider from podium contention, modeling crash likelihood separately is critical.
- Riders with higher historical crash rates are assigned a higher probability of race-ending failure during simulation.

**Anticipated Track Conditions:** Most anticipated track conditions based of the past six Seattle Supercross events.
- This will reflect the expected Seattle race environment, including moisture levels, indoor stadium conditions, and surface slickness.
- These factors influence overall race variance and crash likelihood and are applied globally to all riders within a simulation.

**POF (Probability of Failure):** Each rider's probability of DNF or a major crash.
- This is a modeled probability that a rider experiences a DNF or major crash during the event.
- This probability is derived from historical crash rates and adjusted for anticipated track conditions.
- During each simulation run, riders undergo a probabilistic failure check before final rankings are assigned.

**Rider Demographic:** Each rider's age, fitness level, injury history, etc.
- This variable incorporated features such as age, injury history, and recent fitness status provide additional context for performance variability.

**Conditional Performance:** Each's riders performance history is the riding conditions anticipated in the **Anticipated Track Conditions** (above).
- This varaible captures how riders historically perform under conditions similar to those anticipated in Seattle (Anticipated Track Conditions).
- This allows the model to weight past results more heavily when they occurred under comparable conditions, improving contextual relevance.

#### Simulation Procedure

For each simulation iteration:

- Each rider is assigned a performance score based on base performance and 2025 recent form.

- Seattle-specific and conditional adjustments are applied.

- A probabilistic failure (DNF/crash) check is performed.

- Random race variance is added to reflect uncontrollable race dynamics.

- Riders are ranked by final performance score.

- The Top 3 finishers are recorded.

- This process is repeated 500 times, producing a distribution of podium outcomes rather than a single prediction.
